```python
#Bar Graph for top 20 jobs from our dataset
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 8))                       # set the figure size
plt.title("Job Titles")                           # giving the title for our graph
t1 = df["Title"].value_counts()                   # get the value counts of the "Title" column
subset = t1[:20]                                  # select the top 10 job titles by value count
VR=subset.plot(kind="bar", width=0.8)             # create the bar plot
plt.xticks(fontsize=12)                           # set the x-axis tick labels font size
# add the counts on top of the bars
for i, v in enumerate(subset):
    VR.text(i, v+1, str(v), color='red', fontweight='bold', ha='center')
plt.show()
```

```python
#Bar Graph for top 20 Locations of jobs from our dataset
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))                       # set the figure size
plt.title("Job Locations")                        # giving the title for our graph
t2 = df["Location"].value_counts()                # get the value counts of the "Title" column
subset = t2[:20]                                  # select the top 20 locations by value count
VR=subset.plot(kind="bar", width=0.8)             # create the bar plot
plt.xticks(fontsize=12)                           # set the x-axis tick labels font size
# add the counts on top of the bars
for i, v in enumerate(subset):
    VR.text(i, v+2, str(v), color='red', fontweight='bold', ha='center')
plt.show()
```

```python
#Histogram for Jobs Vs Locations - Top 10
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))  

# select the top 10 job titles by count
top_jobs = df['Title'].value_counts().head(10).index

# select the top 10 locations by count
top_locations = df['Location'].value_counts().head(10).index

# filter the dataframe to include only the top 10 job titles and locations
df_top = df[df['Title'].isin(top_jobs) & df['Location'].isin(top_locations)]

# create the histogram
sns.histplot(data=df_top, x='Title', y='Location', color='blue')

# set the title
plt.title('Histogram of Top 10 Job Titles and Top 10 Locations')

# set the axis labels
plt.xlabel('Job Title')
plt.ylabel('Location')

plt.xticks(rotation='vertical', fontsize=12)

# show the plot
plt.show()
```

```python
#Histogram for Jobs Vs Companies
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))  

# select the top 10 job titles by count
top_jobs = df['Title'].value_counts().head(35).index

# select the top 10 Company by count
top_companies = df['Company'].value_counts().head(25).index

# filter the dataframe to include only the top 10 job titles and top_company
df_top = df[df['Title'].isin(top_jobs) & df['Company'].isin(top_companies)]

# create the histogram
sns.histplot(data=df_top, x='Title', y='Company', color='red')

# set the title
plt.title('Histogram of Top 10 Job Titles and Top 10 Company')

# set the axis labels
plt.xlabel('Job Title')
plt.ylabel('Company')

plt.xticks(rotation='vertical', fontsize=12)

# show the plot
plt.show()
```

```python
#Scatterplot for Jobs Vs Location

# select the top 10 job titles by count
top_jobs = df['Title'].value_counts().head(10).index

# create a list of colors for each job title
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan']

# plot each job title and its corresponding locations as a scatter plot
for i, job in enumerate(top_jobs):
    df_job = df[df['Title'] == job]
    plt.scatter(df_job['Location'], df_job['Title'], color=colors[i], label=job)

# set the title and axis labels
plt.title('Scatter Plot of Top 10 Job Titles and Their Locations')
plt.xlabel('Location')
plt.ylabel('Job Title')

# # set the legend
plt.legend(loc='center left', bbox_to_anchor=(1.05, 0.5))

# rotate the x-axis tick labels to vertical
plt.xticks(rotation='vertical')

# show the plot
plt.show()
```
